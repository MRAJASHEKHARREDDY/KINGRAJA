```python
def upperlower(str):
    upper_count = 0
    lower_count = 0
    for i in str:
        if i.isupper():
            upper_count += 1
        else:
            lower_count += 1
    print(f'No.of upper case characters:{upper_count}')
    print(f'No.of lower case characters:{lower_count}')
upperlower('The quick Brow Fox')
```
